```python
import numpy as np

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = x

colors =['green']*(len(x) - 1)
colors.append('red')

plt.figure()
plt.scatter(x, y, s=100, c=colors)
```

```python
zip_generator = zip([1, 2, 3, 4, 5], [5, 7, 8, 9, 10])
x, y = zip(*zip_generator)
print(x)
print(y)
```

```python
plt.figure()
plt.scatter(x[:2], y[:2], s=100, c='red', label='Abu')
plt.scatter(x[2:], y[2:], s=100, c='blue', label='Alex')
```

```python
plt.xlabel('Whatever')
plt.ylabel('Power 1')
plt.title('Relationship between Alex and Abu')
plt.legend(loc=0, frameon=False, title='legend')
```

# Line charts:

```python
import numpy as np

linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])
q_data = linear_data**2

plt.figure()
plt.plot(linear_data, '-o', q_data, '-o')
```

```python
plt.plot([22, 44, 55], '--r')
```

```python
plt.xlabel('some data')
plt.ylabel('some othe kind of data')
plt.title('A title')
plt.legend(['first', 'second', 'third'], loc=0)
```

```python
plt.fill_between?
```

```python
plt.fill_between(range(len(linear_data)), linear_data, q_data, facecolor='blue', alpha='0.25')
```

```python
plt.figure()

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
plt.plot(observation_dates, linear_data, '-o', observation_dates, q_data, '-o')
```

```python
import pandas as pd
plt.figure()

observation_dates = np.arange('2017-01-01', '2017-01-09', dtype='datetime64[D]')
observation_dates = list(map(pd.to_datetime, observation_dates))

plt.plot(observation_dates, linear_data, '-o', observation_dates, q_data, '-o')
```

```python
x = plt.gca().xaxis

for item in x.get_ticklabels():
    item.set_rotation(45)
```
