# Histogram

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0, scale=1, size = sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n = {}'.format(sample_size))
plt.tight_layout()
```

```python
plt.figure()
Y = np.random.normal(loc=0, scale=1, size=10000)
X = np.random.random(size=10000)
plt.scatter(X, Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1: , 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
lower_right.scatter(X, Y)
top_histogram.hist(X, bins =100)
s = side_histogram.hist(Y, bins=100, orientation='horizontal', normed=True)
side_histogram.invert_xaxis()
```

# Box and Whisker Plots

```python
normal_sample = np.random.normal(loc=0, scale=1, size =10000)
random_sample = np.random.random(size=10000)
gamma_sample = np.random.gamma(2, size=10000)
```

```python
df = pd.DataFrame({'normal': normal_sample, 
                  'random': random_sample,
                  'gamma': gamma_sample})
```

```python
df.describe()
```

```python
df.shape
```

```python
plt.figure()

test = plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')
```

```python
plt.figure()
his = plt.hist(df['gamma'], bins =100)
```

```python
import mpl_toolkits.axes_grid1.inset_locator as mpl_il

plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']], whis='range')

ax2 = mpl_il.inset_axes(plt.gca(), width ='60%', height='40%', loc=2)
ax2.hist(df['gamma'], bins=100)
ax2.margins(x=0.5)
```

```python
ax2.yaxis.tick_right()
```

```python
plt.figure()
plt.boxplot([df['normal'], df['random'], df['gamma']])
```

# Animations

```python
import matplotlib.animation as animation
import numpy as np

n = 100
x = np.random.randn(n)

def update(curr):
    if curr == n:
        a.event_soure.stope()
    plt.cla()
    
    bins = np.arange(-4, 4, 0.5)
    plt.hist(x[:curr], bins=bins)
    plt.axis([-4, 4, 0, 30])
    plt.gca().set_title('Sampling the Normal distribution')
    plt.gca().set_ylabel('Frequency')
    plt.gca().set_xlabel('Value')
    plt.annotate('n={}'.format(curr), [3, 27])
```

```python
fig = plt.figure()
a = animation.FuncAnimation(fig, update, interval=100)
```
