# Reading and Writing CSV files

<br>
Let's import verified_transactions data from TMNT project in CSV format.

```python
import csv

#Setting floating point precisoin. 
%precision 2

# DictReader Creates an object which operates like a regular reader but maps the information read
# into a dict whose keys are given by the optional fieldnames parameter

with open('verified_transactions_test.csv') as csvfile:
     vt = list(csv.DictReader(csvfile))
    
vt[:3] # The first three dictionaries in our list.
```

<br>
`csv.Dictreader` has read in each row of our csv file as a dictionary. `len` shows that our list is comprised of 234 dictionaries.

```python
len(vt)
```

<br>
`keys` gives us the column names of our csv.

```python
list(vt[0].keys())
```

<br>
This is how to find the average amount spent across all verified_transaction dataset. **All values in the dictionaries are strings, so we need to convert to float.**

```python
sum(float(d['price']) for d in vt) / len(vt) # more details
```

```python
vt[0].keys();
```

```python
list(vt[0].values())
```

<br>
Use `set` to return the unique values for the product the verified transactions in our dataset have.

```python
products = set(d['product'] for d in vt)
products
```

<br>
Here's a more complex example where we are grouping the vt by number of products, and finding the average price vt for each group.

```python
PricevtByProd = []

for p in products: # iterate over all the prodcuts
    sumvt = 0
    prodtypecount = 0
    for d in vt: # iterate over all dictionaries in the vt dataset
        if d['product'] == p: # if the product name matches,
            sumvt += float(d['price']) # add the price
            prodtypecount += 1 # increment product count
    PricevtByProd.append((p, sumvt / prodtypecount)) # append the tuple ('product', 'avg price')

PricevtByProd.sort(key=lambda x: x[0])
PricevtByProd
```

<br>
Use `set` to return the unique values for the class types in our dataset.

```python
countries = set(d['country'] for d in vt) # what are the unique countries
countries
```

<br>
And here's an example of how to find the average hwy mpg for each class of vehicle in our dataset.

```python
PriceVtByCountry = []

for c in countries: # iterate over all the countries 
    sumvt = 0
    Countrycount = 0
    for d in vt: # iterate over all dictionaries
        if d['country'] == c: # if the country matches,
            sumvt += float(d['price']) # add the price
            Countrycount += 1 # increment the count
    PriceVtByCountry.append((c, sumvt / Countrycount)) # append the tuple ('Country', 'avg price')

PriceVtByCountry.sort(key=lambda x: x[0])
PriceVtByCountry
```

## Copying Data

Be careful with copying and modifying arrays in NumPy!


`r2` is a slice of `r`

```python
r2 = r[:3,:3]
r2
```

<br>
Set this slice's values to zero ([:] selects the entire array)

```python
r2[:] = 0
r2
```

<br>
`r` has also been changed!

```python
r
```

<br>
To avoid this, use `r.copy` to create a copy that will not affect the original array

```python
r_copy = r.copy()
r_copy
```

<br>
Now when r_copy is modified, r will not be changed.

```python
r_copy[:] = 10
print(r_copy, '\n')
print(r)
```

### Iterating Over Arrays

Let's create a new 4 by 3 array of random numbers 0-9.

```python
test = np.random.randint(0, 10, (4,3))
test
```

<br>
Iterate by row:

```python
for row in test:
    print(row)
```

<br>
Iterate by index:

```python
for i in range(len(test)):
    print(test[i])
```

<br>
Iterate by row and index:

```python
for i, row in enumerate(test):
    print('row', i, 'is', row)
```

<br>
Use `zip` to iterate over multiple iterables.

```python
test2 = test**2
test2
```

```python
for i, j in zip(test, test2):
    print(i,'+',j,'=',i+j)
```
