#### How about writing a Python code:

```python
import pandas as pd
import numpy as np
from scipy import stats
import scipy.special
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n , p= 100, 0.5
k = np.arange(0, 101)
binomial_test = stats.binom.pmf(k, n, p)
binomial_test
```

```python
plt.plot(k, binomial_test, 'o-')
plt.title('Binomial: n = {}, p = {:.2f}'.format(n,p), fontsize=15)
plt.xlabel('Number of heads', fontsize=15)
plt.ylabel('Probability of successes', fontsize=15)
```

```python
# Combinations
scipy.special.binom(10, 2)
```

```python
n , p= 100, 0.2
k = np.arange(0, 101)
binomial_test = stats.binom.pmf(k, n, p)

plt.plot(k, binomial_test, 'o-')
plt.title('Binomial: n = {}, p = {:.2f}'.format(n,p), fontsize=15)
plt.xlabel('Number of heads', fontsize=15)
plt.ylabel('Probability of successes', fontsize=15)
```

```python
norm = 'D:/Python_data_analysis/Applied_python/Introduction_to_python/Ludia_Training/Session_4/Normal_distribution.png'
Image(norm)
```

```python
mu = 0 # mean
sigma = 1 # standard deviation

x= np.arange(-5, 5, 0.1)
y = stats.norm.pdf(x, mu, 1)

plt.plot(x, y)
plt.title('Normal: $mu$={:.1f},   $sigma^2$={:.1f}'. format(mu, sigma))
plt.xlabel('x')
plt.ylabel('Probability')
```

Formula for standard deviation
$$\sqrt{\frac{1}{N} \sum_{i=1}^N (x_i - \overline{x})^2}$$

```python
np.random.normal?
```

```python
distribution = np.random.normal(0, size=1000)

np.sqrt(np.sum((np.mean(distribution)-distribution)**2)/len(distribution))
```

```python
np.std(distribution)
```

```python
import scipy.stats as stats
stats.kurtosis(distribution)
```

```python
stats.skew(distribution)
```
